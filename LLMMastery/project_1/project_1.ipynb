# Project 1: Simple LLM

In this project, I implement a basic decoder-only transformer model. 

1. Implementation:
   - Single attention head
   - Basic positional encoding
   - 6 transformer layers
   - Context window: 512 tokens

2. Data Processing:
   - Character-level tokenization
   - Wikipedia subset (2GB)
   - Basic cleaning pipeline
   - Simple batching

3. Training:
   - Single GPU training
   - Basic Adam optimizer
   - Fixed learning rate
   - Cross-entropy loss

4. Evaluation:
   - Perplexity
   - Basic text generation
   - Word prediction
